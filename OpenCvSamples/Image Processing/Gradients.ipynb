# Gradients
Understanding gradients will allow us to eventually understand edge detection which we will use later on, since its an important aspect of object detection in general. Make sure to watch the lecture video for the full overview!

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# o for black and white
img=cv2.imread('Computer-Vision-with-Python/DATA/sudoku.jpg',0)
type(img)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
# With sobel x we can see the edges in the y-axis clearly
display_img(sobelx)
```

```python
# With sobely we can see the edges in the x axis clearly
display_img(sobely)
```

```python
merge=cv2.addWeighted(sobelx,0.3,sobely,0.7,0)
```

```python
#Here we can see edges from both the axes to an extent
display_img(merge)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

## Morphological Operators


```python
kernel = np.ones((4,4),np.uint8)
gradient=cv2.morphologyEx(merge,cv2.MORPH_GRADIENT,kernel)
```

```python
#Note that we can see the edges alone very clearly
display_img(gradient)
```
