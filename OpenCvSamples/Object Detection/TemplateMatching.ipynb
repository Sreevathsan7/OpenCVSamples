```python
#All 6 methods for comparison
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
height,width,channels=face.shape

for m in methods:
    #Making a copy of the image
    full_copy=full.copy()
    
    #Using eval
    method=eval(m)
    
    #Applying template matching with that method
    res=cv2.matchTemplate(full_copy,face,method)
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    #Obtaining the points of the rectangle
    
     # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
    # Notice the coloring on the last 2 left hand side images.
    
    for method in ['cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']:
        top_left=min_loc
    else:
        top_left=max_loc
   

    bottom_right = (top_left[0] + width, top_left[1] + height)  
    
    cv2.rectangle(full_copy,top_left,bottom_right,(0,255,0),thickness=10)
    
    #ploting the images
    #Here 121 means 1 row 2 columns and 1st image
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Result of Template Matching')
    
    #Here 122 means 1 row 2 columns and 2nd image
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detected Point')
    plt.suptitle(m)
    
    #print blank spaces so that plt doesnt overwrite
    plt.show()
    
    print('\n')
    print('\n')
    
```
